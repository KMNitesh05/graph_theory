```python
import networkx as nx
import matplotlib.pyplot as plt 
from sklearn.cluster import KMeans
import numpy as np
#from mayavi import mlab
```

```python

# Create a random graph with 20 nodes and 30 edges
G = nx.gnm_random_graph(100, 300)

# Print the number of nodes and edges in the graph
print(f"Number of nodes: {G.number_of_nodes()}")
print(f"Number of edges: {G.number_of_edges()}")

# Calculate the PageRank of each node in the graph
pr = nx.pagerank(G)

# Print the PageRank values for each node
#for node, pagerank_value in pr.items():
    #print(f"Node {node} has a PageRank of {pagerank_value}")


# Create a histogram of the PageRank values
plt.hist(list(pr.values()), bins=20)
plt.show()


# Create a list of colors based on the PageRank values
colors = [plt.cm.viridis(pagerank_value) for pagerank_value in pr.values()]

# Draw the graph and color each node based on its PageRank value
nx.draw(G)#, node_color=colors, with_labels=True)
plt.show()

```

```python
# Calculate the PageRank of each node in the graph
pr = nx.pagerank(G)

# Reshape the PageRank values into a 2D array
pr_values = np.array(list(pr.values())).reshape(-1, 1)

# Perform K-means clustering on the PageRank values
kmeans = KMeans(n_clusters=10).fit(pr_values)

# Print the cluster labels for each node
#for node, label in zip(pr.keys(), kmeans.labels_):
    #print(f"Node {node} is in cluster {label}")

# Create a histogram of the cluster sizes
plt.hist(np.bincount(kmeans.labels_), bins=5)
plt.show()
```

```python
# Print the cluster number and the nodes within each cluster
for label in range(10):
    cluster_nodes = [node for node, node_label in zip(pr.keys(), kmeans.labels_) if node_label == label]
    print(f"Cluster {label}: {cluster_nodes}")
```

```python
# Print the cluster number and the number of nodes in each cluster
for cluster_num in range(10):
    nodes_in_cluster = np.sum(kmeans.labels_ == cluster_num)
    print(f"Cluster {cluster_num} has {nodes_in_cluster} nodes")
```

```python
# Create a list of colors based on the cluster labels
colors = [plt.cm.tab10(label) for label in kmeans.labels_]

# Draw the graph and color each node based on its cluster label
nx.draw(G, node_color=colors, with_labels=True,node_size=100)
plt.show()
```

```python
# Draw the graph using the spring layout algorithm
pos = nx.spring_layout(G)
nx.draw(G, pos=pos, with_labels=False,node_color=colors,node_size=200)
```

```python
# Compute the probability of each cluster
num_nodes = G.number_of_nodes()
clust_prob = []
cluster_size = []
for cluster_num in range(10):
    nodes_in_cluster = np.sum(kmeans.labels_ == cluster_num)
    cluster_probability = nodes_in_cluster / num_nodes
    clust_prob.append(cluster_probability)
    cluster_size.append(nodes_in_cluster)
    print(f"Cluster {cluster_num} has a probability of {cluster_probability}")
```

```python
clusters = np.arange(0, 10, 1).tolist()
plt.plot(clusters,clust_prob)
```

$$R(\tau) = \frac{\sum_{t=1}^{N-\tau} (x_t - \mu)(x_{t+\tau} - \mu)}{\sum_{t=1}^N (x_t - \mu)^2}$$



```python
import numpy as np

def autocorrelation(x):
    # Compute the mean of the signal
    mu = np.mean(x)

    # Initialize the autocorrelation function with zeros
    autocorr = np.zeros(len(x))

    # Compute the autocorrelation function
    for tau in range(len(x)):
        autocorr[tau] = np.sum((x[:len(x)-tau] - mu)*(x[tau:] - mu)) / np.sum((x - mu)**2)

    return autocorr

# Generate a list of 100 random numbers
x = np.random.randn(100)
auto
# Compute and print the autocorrelation function
print(autocorrelation(x))
plt.plot(autocorrelation(x))
```
