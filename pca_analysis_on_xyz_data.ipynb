## PCA on XYZ data ## 

```python
import random
import matplotlib.pyplot as plt

# Generate 1000 random x, y, and z values
x = [random.random() for _ in range(1000)]
y = [random.random() for _ in range(1000)]
z = [random.random() for _ in range(1000)]

# Create the scatterplot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z)

# Show the plot
plt.show()
```

```python
import numpy as np
from sklearn.decomposition import PCA

# Define the dataset (replace with your own data)
data = np.array([x,y,z,])

# Perform PCA on the data
pca = PCA(n_components=2)
pca.fit(data)

# Get the PC1 and PC2 values
pc1 = pca.components_[0]
pc2 = pca.components_[1]

# Plot the PC1 and PC2 values
plt.scatter(pc1, pc2)
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.show()
```
