```python
import random
import numpy as np 
import matplotlib.pyplot as plt 
import networkx as nx
from mayavi import mlab
#from simplicial import simplicial_complex
```

```python
# Generate a random graph with 20 nodes and 60 edges using networkx
G = nx.gnm_random_graph(20, 60)

# Draw the graph using matplotlib
nx.draw(G)

# Show the plot
plt.show()
```

```python
# Find all the cliques in the graph
cliques = list(nx.find_cliques(G))

# Create the clique complex by taking the subsets of the cliques
C = [set(c) for c in cliques]

# Save the 0-simplices, 1-simplices, 2-simplices, and 3-simplices of the clique complex to a list
simplices = [
    [s for s in C if len(s) == k]
    for k in range(4)
]

# Print the number of 0-simplices, 1-simplices, 2-simplices, and 3-simplices in the clique complex
print(f"Number of 0-simplices: {len(simplices[0])}")
print(f"Number of 1-simplices: {len(simplices[1])}")
print(f"Number of 2-simplices: {len(simplices[2])}")
print(f"Number of 3-simplices: {len(simplices[3])}")

# Draw the graph and the clique complex using matplotlib
nx.draw(G, with_labels=True)
#for simplex in C:
    #plt.plot(list(simplex), '-o')
#plt.show()


```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Generate a random graph with 20 nodes and 60 edges
G = nx.gnm_random_graph(20, 60)

# Find all the cliques in the graph
cliques = list(nx.find_cliques(G))

# Create the clique complex by taking the subsets of the cliques
C = [set(c) for c in cliques]

# Save the 0-simplices, 1-simplices, 2-simplices, and 3-simplices of the clique complex to a list
simplices = [
    [s for s in C if len(s) == k]
    for k in range(4)
]

# Draw the graph and the clique complex using matplotlib
nx.draw(G, with_labels=True)
for simplex in C:
    # Draw the edges in the current simplex as lines between the nodes
    nx.draw_networkx_edges(G, pos=nx.spring_layout(G), edgelist=list(simplex), edge_color="red")
plt.show()
```
