```python
import networkx as nx
import community

# Generate the graphs and find the communities in each
graphs = []
communities = []
for i in range(100):
    
    ## you can add your graphs here ## 
    
    G = nx.Graph()
    for j in range(100):
        G.add_node(j)
    for k in range(200):
        u = random.randint(0,99)
        v = random.randint(0,99)
        while u == v:
            v = random.randint(0,99)
        G.add_edge(u,v)
    graphs.append(G)
    communities.append(community.best_partition(G))

#print(graphs)
# Compare the communities in consecutive pairs of graphs
# and compute the probability of survival
survival_probability = []
for i in range(99):
    G1 = graphs[i]
    G2 = graphs[i+1]
    c1 = communities[i]
    c2 = communities[i+1]
    survival_count = 0
    for node in G1.nodes():
        #print(node)
        #print(c1[node])
        if c1[node] == c2[node]:
            survival_count += 1
    survival_prob = survival_count / G1.number_of_nodes()
    print("Probability of survival from graph", i+1, "to graph", i+2, ":", survival_prob)
    survival_probability.append(survival_prob)
```

```python
def autocorrelation(x):
    # Compute the mean of the signal
    mu = np.mean(x)

    # Initialize the autocorrelation function with zeros
    autocorr = np.zeros(len(x))

    # Compute the autocorrelation function
    for tau in range(len(x)):
        autocorr[tau] = np.sum((x[:len(x)-tau] - mu)*(x[tau:] - mu)) / np.sum((x - mu)**2)

    return autocorr

# Compute and print the autocorrelation function
print(autocorrelation(survival_probability))
plt.plot(autocorrelation(survival_probability))
```

```python
# Compute the autocorrelation function and its Fourier transform
autocorr = autocorrelation(survival_probability)
fourier = np.fft.fft(autocorr)

# Print the Fourier transform of the autocorrelation function
print(fourier)

plt.plot(fourier)
plt.xlabel("frequency")
plt.ylabel("Intensity")
```
